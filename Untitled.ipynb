```python
import pandas as pd
import numpy as np

df = pd.read_excel('schedule.xlsx',skiprows=3)

df = df.iloc[:,1:6]
df.rename(columns={'Unnamed: 1':'date','TIME ':'time','Unnamed: 3':'subject','STAFF INCHARGE':'staff',
                   'Online platform contact link':'links'},inplace=True)

df.fillna(method='ffill',inplace=True)
df = df.iloc[:-5,:]
df.staff = df.staff.str.lower()
df.staff.replace('[a-z]+\.','',regex=True,inplace=True)
df.staff.replace('r','asha',inplace=True)
my_lec = {'saraswathi':'Ethics','gowri':'Web Programming','aarthi':'Software Quality Assurance',
          'prabhavathi':'Wireless Sensor Networks','asha':'Artificial Intelligence'}

df.links.replace('\\\n','',regex=True,inplace=True)


#one more last task to rename staff according to my lecturers
def func(x):
    for nm in my_lec.keys():
        if nm in x:
            return nm
    return np.nan

df.staff = df.staff.apply(func)
df.dropna(inplace=True)

df.date = df.date.apply(lambda x:x.split('.')[0])
df.time = df.time.apply(lambda x:x.split('-')[0].split('.')[0])
df.to_csv('sch.csv',index=False)
df
```
